```python
import numpy as np
import matplotlib.pyplot as plt
import japanize_matplotlib
import seaborn as sns
```

```python
"""
データを読み込み
"""

# csvファイルの読み込みと空のリストに追加
import pandas as pd
df = pd.read_csv(filepath_or_buffer=r"/Users/s.ogura/Documents/CRLEA/data/20220422/kokusuei_0422.csv",sep=",")
df = df.replace({'科目': {10: '国語', 20: '数学', 30: '英語'}})
df.head(3)
```

```python
df= df[~df["想定ペース"].isna()]
```

```python
# dataframe 分ける
df_kamoku = df.groupby(df.科目)
jap_df = df[df["科目"] == "国語"]
math_df= df[df["科目"] == "数学"]
eng_df= df[df["科目"] == "英語"]
```

```python
#グラフ描画
def graph_test(df):
    x = np.linspace(0,1000,1000)
    y = x
    plt.plot(x,y, color = "black", ls = "--")
    plt.title("想定ペース vs 実績ペース",
                      fontsize=20) # タイトル
    plt.xlabel("想定ペース", fontsize=20) # x軸ラベル
    plt.ylabel("実績ペース", fontsize=20) # y軸ラベル
    plt.grid(True) # 目盛線の表示
    plt.tick_params(labelsize = 12) # 目盛線のラベルサイズ
    plt.xticks(np.arange(0, 1001, step=100))
    plt.yticks(np.arange(0, 1001, step=100))
    sns.scatterplot(x='想定ペース', y='ペース平均', hue='科目', data=df)
```

```python
#グラフ描画
def graph_test(df,title="想定ペース vs 実績ペース"):
    plt.figure(figsize=(5,5))
    x = np.linspace(0,1000,1000)
    y = x
    plt.plot(x,y, color = "black", ls = "--")
    plt.title(title,
                      fontsize=20) # タイトル
    plt.xlabel("想定ペース", fontsize=20) # x軸ラベル
    plt.ylabel("実績ペース", fontsize=20) # y軸ラベル
    plt.grid(True) # 目盛線の表示
    plt.tick_params(labelsize = 12) # 目盛線のラベルサイズ
    plt.xticks(np.arange(0, 1001, step=100))
    plt.yticks(np.arange(0, 1001, step=100))
    sns.scatterplot(x='想定ペース', y='ペース平均', hue='科目',alpha = 0.5, data=df)
```

```python
#グラフ描画
graph_test(df)
```

```python
graph_test(jap_df,"国語")
```

```python
graph_test(math_df,"数学")
```

```python
graph_test(eng_df,"英語")
```
